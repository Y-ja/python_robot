```python
import numpy as np
from pop import Cds, delay
from pop import Pilot

car = Pilot.AutoCar()
cds = Cds(7)

try:
    while True:
        value = cds.readAverage()  # 조도 값을 읽어오기
        print(f"조도 값: {value}")

        # 조도 값을 기반으로 자동차 속도 세분화
        if value >= 700:
            speed = 100  # 최대 속도
        elif value >= 500:
            speed = 75
        elif value >= 300:
            speed = 50
        elif value >= 100:
            speed = 25
        else:
            speed = 0  # 어두워지면 속도 0

        car.setSpeed(speed)  # 자동차 속도 설정

        # 조도 값에 따라 자동차 동작 결정
        if value >= 100:
            car.forward()  # 밝으면 전진
        else:
            car.stop()  # 어두워지면 정지
            print("어두워졌습니다. 자동차가 정지합니다.")

        delay(500)  # 500 밀리초 대기

except KeyboardInterrupt:
    car.stop()  # 정지 버튼을 누르면 즉시 멈춤
    print("프로그램이 종료되었습니다.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
import numpy as np
from pop import Cds, delay
from pop import Pilot

car = Pilot.AutoCar()
cds = Cds(7)

try:
    while True:
        value = cds.readAverage()  # 조도 값을 읽어오기
        print(f"조도 값: {value}")

        if value < 50:
            print("주의: 조도가 너무 낮습니다!")  # 알림 출력

        # 조도 값을 기반으로 자동차 속도 설정
        speed = value / 16  # 조도 값을 16으로 나눠 속도로 설정 (0~100의 값으로)
        car.setSpeed(speed)  # 자동차 속도 설정

        # 조도 값에 따라 자동차 동작 결정
        if value >= 700:
            car.forward()  # 밝으면 전진
        elif value >= 50:
            car.forward()  # 중간 밝기에서 전진
        else:
            car.stop()  # 어두워지면 정지
            print("어두워졌습니다. 자동차가 정지합니다.")

        delay(500)  # 500 밀리초 대기

except KeyboardInterrupt:
    car.stop()  # 정지 버튼을 누르면 즉시 멈춤
    print("프로그램이 종료되었습니다.")
except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```
