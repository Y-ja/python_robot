```python
# 선형 회귀
from pop import AI  # AI 모듈에서 Linear_Regression 클래스를 가져옵니다.

# 선형 회귀 객체 생성
LR = AI.Linear_Regression()  # Linear_Regression 클래스의 인스턴스를 생성합니다.
```

```python
# 입력 데이터(키)
LR.X_data = [[173], [171], [162], [181], [157], [169], [177], [159], [182]]

# 정답 데이터(발 크기)
LR.Y_data = [[270], [275], [245], [280], [230], [265], [270], [250], [275]]
```

```python
import matplotlib.pyplot as plt
plt.plot(LR.X_data, LR.Y_data)
plt.show()
```

```python
plt.plot(LR.X_data, LR.Y_data, 'ro')
plt.show()
```

```python
LR.train()
```

```python
# 데이터 검증(훈련 데이터로 예측값 출력)
pred = LR.run()
```

```python
plt.plot(LR.X_data, pred, 'ro')
plt.show()
```

```python
LR.train(times = 1000, print_every = 100)
```

```python
pred = LR.run()
```

```python
x_pred = ([150], [160], [170], [180], [190])
pred = LR.run(x_pred)
```

```python
plt.plot(x_pred, pred, 'ro')
plt.show()
```

```python
# 선형 회귀
from pop import AI  # AI 모듈에서 Linear_Regression 클래스를 가져옵니다.

ANN = AI.ANN(input_size=3 , output_size=3)
```

```python
# 입력 데이터 (X_data): 3개의 특성을 가진 데이터
ANN.X_data = [
    [73, 90, 1],    # 첫 번째 데이터 포인트
    [62, 70, 0],    # 두 번째 데이터 포인트
    [83, 100, 2],   # 세 번째 데이터 포인트
    [110, 150, 7],  # 네 번째 데이터 포인트
    [139, 220, 15], # 다섯 번째 데이터 포인트
    [123, 190, 10], # 여섯 번째 데이터 포인트
    [177, 275, 7],  # 일곱 번째 데이터 포인트
    [159, 240, 35], # 여덟 번째 데이터 포인트
    [180, 280, 15]  # 아홉 번째 데이터 포인트
]

# 출력 데이터 (Y_data): 각 입력에 대한 클래스 레이블 (원-핫 인코딩)
ANN.Y_data = [
    [1, 0, 0],  # 첫 번째 데이터 포인트: 클래스 1
    [1, 0, 0],  # 두 번째 데이터 포인트: 클래스 1
    [1, 0, 0],  # 세 번째 데이터 포인트: 클래스 1
    [0, 1, 0],  # 네 번째 데이터 포인트: 클래스 2
    [0, 1, 0],  # 다섯 번째 데이터 포인트: 클래스 2
    [0, 1, 0],  # 여섯 번째 데이터 포인트: 클래스 2
    [0, 1, 0],  # 일곱 번째 데이터 포인트: 클래스 2
    [0, 0, 1],  # 여덟 번째 데이터 포인트: 클래스 3
    [0, 0, 1]   # 아홉 번째 데이터 포인트: 클래스 3
]
```

```python
ANN.train()
```

```python
ANN.train(times=1000 , print_every=100)
```

```python
ANN.run([[174,270,10],[57,70,1],[140,220,10]])
```

```python
ANN.train(times=100000 , print_every=10000)
```

```python
# ANN.run 메서드 호출: 입력 데이터로 두 개의 데이터 포인트를 제공합니다.
ANN.run([[140, 220, 10], [140, 200, 11]])
```
